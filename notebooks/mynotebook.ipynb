# Using Jupyter notebooks

This is a jupyter notebook. It's a way to share code in a way that's much more accessible than code alone. A notebook combines 3 elements:

* Normal text, written in Markdown
* Snippets of code, written in Python
* The output of that code, which is generated by the notebook itself when it runs the code

Below I switch from the normal text to some Python code. Underneath that, the notebook shows the output:

```python
print ("Hello world. How boring.")
```

Here's another example: code that creates a **function** and then runs it. You'll then see the output. Note, however, that these lines *follow on* from any other code earlier.

```python
def myfunction(myname):
    print (myname)

myfunction("Paul")
```

## Creating notebooks

There are various ways of creating a notebook. 

One way is to install [Anaconda](https://www.continuum.io/downloads). Once that's installed, run it, and then click on the **Launch** button for Jupyter notebook. This should open it in your browser at an address that starts with `http://localhost:8889/notebooks/`.

Another way is to use [SageMathCloud](https://cloud.sagemath.com/projects). This is a web-based interface for creating notebooks: it's slower than Anaconda but you can get started with it without having to install anything.

## Sharing notebooks publicly

Once created you can share your notebook using [nbviewer](https://nbviewer.jupyter.org/).

To do that:

1. First, download your notebook from Jupyter by clicking on **File > Download as > Notebook (ipynb)**
2. Create a repository on Github (or if you already have one for your project, go there) and upload that ipynb file. Copy the link to that file into
3. Go to [nbviewer](https://nbviewer.jupyter.org/) and paste the URL into the box on the homepage. You should now be given a link to the readable public version.
